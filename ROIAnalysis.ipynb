```python
import os

images = os.listdir('./pred_mask')
```

```python
from PIL import Image

```

```python
import matplotlib.pyplot as plt 
```

```python
name = images[20]
img = Image.open(f'./train_data/Images/' + name)
mask = Image.open(f'./train_data/Masks/' + name)
pred_mask = Image.open(f'./pred_mask/' + name)

# Plot the input image, ground truth and the predicted output
plt.figure(figsize=(10,10));
plt.subplot(131);
plt.imshow(img);
plt.title('Image')
plt.axis('off');
plt.subplot(132);
plt.imshow(mask);
plt.title('Ground Truth')
plt.axis('off');
plt.subplot(133);
plt.imshow(pred_mask);
plt.title('Segmentation Output')
plt.axis('off');
```
